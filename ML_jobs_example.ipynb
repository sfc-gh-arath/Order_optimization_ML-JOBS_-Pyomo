```
from snowflake.ml.jobs import submit_file, submit_directory, submit_from_stage

compute_pool = "MYPOOL"

job3 = submit_from_stage(
    "@job_files/scripts/",
    compute_pool,
    entrypoint="@job_files/scripts/my_script.py",
    stage_name="payload_stage",
    args=["aswinee", "rath"], 
)
```

```
result = job3.result() # Hello world
print(result)
```
